### PostgreSQL with Python

For this chapter, we will be using the package 'pyscopg'. We can install it by running the following code in the command prompt:

python -m pip install psycopg2-binary

```python
import psycopg2 as pg2
```

Now we connect it to the database.

```python
secret = 'Your password'
```

```python
conn = pg2.connect(
    host="localhost",
    port=5432,
    database="dvdrental",
    user="postgres",
    password="secret"
)
```

The next step is to obtain the cursor. It is the control structure that enables to work with records in the database.

```python
cur = conn.cursor()
```

```python
cur.execute('SELECT * FROM payment')
```

```python
### The cur gives us several options. Let's try to show the first record.

cur.fetchone()
```

```python
### We can use fetchmany to display the first 5 records.

cur.fetchmany(5)
```

```python
data = cur.fetchmany(5)
```

```python
data[0]
```

Once complete, we should close the connection.

```python
conn.close()
```
